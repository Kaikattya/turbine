# 1. Построение процесса расширения пара в турбине.
## 1.1 Определение расходов пара на входе в турбину $G_0$ и в конденсатор $G_к$.

# Из графика получаем, что коэффициент равен 0.89

![image.png]( image.png)

```python
import class11 as rasch
import class12 as graf
```

```python
MPa = 10 ** 6
kPa = 10 ** 3
MW  = 10 ** 6
ch = rasch.CalcExpenditurePar(p_o = 22 * MPa,
                 t_o = 600, 
                 p_nn = 4 * MPa, 
                 t_nn = 600, 
                 p_k = 3.5 * kPa, 
                 effect_oi = 0.85, 
                 effect_mex = 0.994, 
                 effect_eg = 0.99,
                 n = 330 * MW,
                t_nv = 260)
g_0, g_k = ch.calc_expenditure()
print(f"Расход свежего пара перед турбиной G_o = {g_0}")
print(f"Расход пара в конденсаторе G_k = {g_k}")
```

```python
fig, ax  = graf.plt.subplots(1, 1, figsize=(15, 15))
graf.plot_hs_diagram(
    ax,
    points=[ch.o_teor, ch.o, ch._1_teor, ch._1_, ch.nn_teor, ch.nn, ch.k, ch.k_teor]
)
graf.plot_process(ax, points=[ch.o_teor, ch.o, ch._1_], color='black')
graf.plot_process(ax, points=[ch.nn_teor, ch.nn, ch.k], color='black')
graf.plot_process(ax, points=[ch.o_teor, ch.o, ch._1_teor], alpha=0.5, color='grey')
graf.plot_process(ax, points=[ch.nn_teor, ch.nn, ch.k_teor], alpha=0.5, color='grey')
```
